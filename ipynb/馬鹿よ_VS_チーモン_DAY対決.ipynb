#### 以下準備作業

```python
import sqlite3
conn = sqlite3.connect('bakayo_VS_chimon.db')
c = conn.cursor()
```

```python
import pandas as pd
```

```python
def count_tweet(tbl_name):
    query = '''
    SELECT count(id), month, day, year FROM {tbl} WHERE month = 'Dec' GROUP BY month, day, year
    '''.format(tbl=tbl_name)
    count_id = c.execute(query)
    
    df_tbl = pd.DataFrame(columns=['count_id','month','day','year'])
    
    for row in count_id:
        print(row)
        df_addition_row = pd.DataFrame([[row[0], row[1], row[2], row[3]]])
        df_addition_row.columns = ['count_id','month','day','year']
        df_tbl = pd.concat([df_tbl,df_addition_row],axis=0)
        
    return df_tbl
```

### 馬鹿よさんのグラフ

```python
df_bakayo = count_tweet('bakayo_time')
```
