#### 以下準備作業

### 馬鹿よさんのグラフ

```python
import sqlite3
conn = sqlite3.connect('bakayo_VS_chimon.db')
c = conn.cursor()
```

```python
query = '''
SELECT count(id), month, day, year FROM bakayo_time GROUP BY month, day, year
'''
count_id = c.execute(query)
```

```python
import pandas as pd
df_bakayo = pd.DataFrame(columns=['count_id','month','day','year'])
```

```python
for row in count_id:
    print(row)
    df_addition_row = pd.DataFrame([[row[0], row[1], row[2], row[3]]])
    df_addition_row.columns = ['count_id','month','day','year']
    df_bakayo = pd.concat([df_bakayo,df_addition_row],axis=0)
```

```python
df_bakayo
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_bakayo.plot.bar(y=['count_id'], x=['day'], color="#F8766D", alpha=.5)
#plt.title(u'棒グラフ', size=16)
```
