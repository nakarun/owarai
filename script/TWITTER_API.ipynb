## TIMELINE LIST

```python
from requests_oauthlib import OAuth1Session
import json
import settings
```

```python
twitter = OAuth1Session(settings.CONSUMER_KEY, settings.CONSUMER_SECRET, settings.ACCESS_TOKEN, settings.ACCESS_TOKEN_SECRET)
```

```python
params = {}
req = twitter.get("https://api.twitter.com/1.1/statuses/home_timeline.json", params = params)
```

```python
timeline = json.loads(req.text)
```

```python
for tweet in timeline:
    print('***')
    print(tweet["text"])
```

## GET SOMEONE'S TWEET

```python
#coding:utf-8
import twitter,random
import settings

api = twitter.Api(settings.CONSUMER_KEY, settings.CONSUMER_SECRET, settings.ACCESS_TOKEN, settings.ACCESS_TOKEN_SECRET)
 
tlAry = api.GetUserTimeline('kanakichiyade',count=100)
tweetlist=[]
 
for s in tlAry:
    if s.text[0] != '@':
        tweetlist.append(s.text)
                                                                                                                           
api.PostUpdates(status=tweetlist[random.randint(0,99)])
```

## FOLLOW LIST

```python
twitter = OAuth1Session(settings.CONSUMER_KEY, settings.CONSUMER_SECRET, settings.ACCESS_TOKEN, settings.ACCESS_TOKEN_SECRET)
params = {}
req = twitter.get("https://api.twitter.com/1.1/friends/list.json", params = params)
friends = json.loads(req.text)
for friend in friends['users']:
    print('***')
    print(friend["name"])
```
